```python
import pandas as pd
import numpy as np
import geopandas as gpd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
filepath = '/Users/kevinmarlis/Developer/LA County PurpleAir/data.geojson'
df_places = gpd.read_file(filepath)
```

```python
f, ax = plt.subplots(1, figsize=(30, 30))
ax = df_places.plot(column='pm_25_weighted_avg', cmap='twilight_shifted', legend=True, ax=ax)
ax.set_xlim([-118.8, -117.6])
ax.set_ylim([33.65, 34.45])

plt.show()
```
